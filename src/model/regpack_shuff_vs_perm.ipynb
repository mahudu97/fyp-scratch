```python
import matplotlib.pyplot as plt
import math
```

```python
MIN_EL = 32
```

```python
def perm_req_reg_dp(num_el):
    req_reg = 8 # permute operands
    req_reg += math.ceil(num_el/8.0)
    return req_reg
def perm_req_reg_sp(num_el):
    req_reg = 16 # permute operands
    req_reg += math.ceil(num_el/16.0)
    return req_reg
```

```python
def shuff_req_reg(num_el):
    req_reg = math.ceil(num_el/4.0)
    return req_reg
```

```python
req_reg_shuff = [shuff_req_reg(i) for i in range(MIN_EL, 121)]
req_reg_perm_dp = [perm_req_reg_dp(i) for i in range(MIN_EL, 177)]
req_reg_perm_sp = [perm_req_reg_sp(i) for i in range(MIN_EL, 225)]
```

```python
plt.figure(dpi=150, figsize=(10,5))
plt.plot(range(MIN_EL, 121), req_reg_shuff, label="VSHUFF64X2 (DP & SP)")
plt.plot(range(MIN_EL, 177), req_reg_perm_dp, label="VPERMD (DP)")
plt.plot(range(MIN_EL, 225), req_reg_perm_sp, label="VPERMD (SP)")
plt.xlabel("Number of Unique Non-Zeros")
plt.ylabel("Number of Registers Required")

plt.xticks([32, 50, 75, 100, 120, 150, 175, 200, 224])
plt.legend()
plt.savefig("./../../plots/reg_packing_comp/regpack_shuff_vs_perm.pdf")
plt.show()
```
